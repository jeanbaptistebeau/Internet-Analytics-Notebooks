```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
from collections import Counter
%matplotlib inline
```

```python
def read_edge_file(pathname):
    """Reads a graph from a csv file"""
    f = open(pathname, 'r')
    G = nx.Graph()
    for line in f:
        if line[0] == '#':
            continue
        line = f.readline().split(',')
        G.add_edge(int(line[0]) , int(line[1]))
    f.close()
    return G
```

```python
G1 = read_edge_file('../data/network1.csv')
```

```python
print("The network has", len(G1), "nodes", "and", G1.size(), "edges")
```

#### Exercise 2.2:  Node degree distribution

```python
def ccdf(graph):
    """ Computes the Complementary Cumulative Distribution 
    of the degree for the provided graph.
    Returns two numpy arrays, the first contains the degrees while
    the second contains the probabilities.
    """
    degrees = nx.degree(graph)
    deg_distr = Counter()
    for deg in degrees.values():
        deg_distr[deg] += 1
    cumulative = np.zeros(len(deg_distr))
    cumulative[0] = deg_distr[0]
    for i in range(1, len(cumulative)):
        cumulative[i] = deg_distr[i] + cumulative[i-1]
    return np.array(list(deg_distr.keys())), 1 - cumulative / len(graph)
```

```python
degs, counts = ccdf(G1)
N = 15

fig, ax = plt.subplots()
x = degs[:N]
y = counts[:N]
plt.loglog(x, y)
ax.set_title("Complementary Cumulative Degree Distribution")
ax.xaxis.set_label_text("Degrees")
ax.yaxis.set_label_text("Cumulative Distribution")
plt.show()
```

A loglog plot is useful for ploting a complementary cumulative distribution function.

This distribution looks like a Power law as the curve is straight. We didn't plot further than degree 15 because the probability drop to 0.01%.

Here we tried to find the slope of the Power law by fitting a function into the data.

```python
from scipy.optimize import curve_fit

power_series = lambda x, a: np.power(x, -a)

popt, pcov = curve_fit(power_series, x, y)
fig, ax = plt.subplots()
ax.semilogy(x, y)
gen_data = np.power(x, -popt[0])
ax.loglog(x, gen_data, color='r')
ax.set_title("Parameter estimation of the Complementary Cumulative Degree Distribution")
ax.xaxis.set_label_text("Degrees")
ax.yaxis.set_label_text("Cumulative Distribution")
plt.show()
print("The parameter of the power series is", popt[0])
```

Network1 seems to be a small world, even smaller than what Duncan Watts exlpains as the average shortest path is $\approx$4.7 links in this graph.

```python
def path_length_distribution(path_lengths):
    """Computes the distribution of the path lengths
    Returns a dictionary where the keys are the lengths
    and the values are the counts of paths having that length.
    """
    distribution = Counter()
    for dest in path_lengths.values():
        for dist in dest.values():
            distribution[dist] += 1
    return {k: distribution[k] / len(path_lengths) for k in distribution.keys()}
```

```python
# x = distance, y = (number of couple of nodes at this distance / number of nodes)  Do we need division?
distribution = path_length_distribution(path_lengths)
```

```python
fig, ax = plt.subplots()
ax.set_title("Average Short Path Distribution")
ax.xaxis.set_label_text("Distance")
ax.yaxis.set_label_text("Nodes")
ax.plot(list(distribution.keys()), list(distribution.values()))
plt.show()
```
